* Submit a new risk model builder request

```python
risk_model_builder.new_request(universe = 'SP500',
                               template = 'default',
                               startDate = '2018-01-31',
                               endDate = '2018-12-31',
                               freq = '1me')
```

* Wait for it to finish

```python
risk_model_builder.wait(max_wait_secs = 600)
```
